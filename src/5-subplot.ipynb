### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 rows, 2 columns

# --- 1st Subplot: Line Chart ---
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]
axs[0, 0].plot(x, y, color='blue', marker='o', linestyle='-')
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")

# --- 2nd Subplot: Bar Chart ---
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# --- 3rd Subplot: Scatter Plot ---
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")

# --- 4th Subplot: Pie Chart ---
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)
axs[1, 1].set_title("Pie Chart")

# Adjust layout for better spacing
plt.tight_layout()

# Show the figure with subplots
plt.show()
```
